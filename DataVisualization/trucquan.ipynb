```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file JSON vào DataFrame
df = pd.read_json("D:/lby/Nam4/HKII/BIGDATA/LeThiBaoYen_Buoi4/hanghoa.json")

# Chuyển kiểu dữ liệu của cột CreateDate sang kiểu datetime
df['CreatedDate'] = pd.to_datetime(df['CreatedDate'], format='%Y-%m-%d %H:%M:%S.%f')

# Lấy số lượng hàng hóa được tạo trong mỗi ngày
product_count = df.groupby(df['CreatedDate'].dt.date).size()

# Vẽ biểu đồ số lượng hàng hóa được tạo trong từng ngày
plt.xticks(rotation= 90)

plt.plot(product_count.index, product_count.values)
plt.xlabel('Ngày')
plt.ylabel('Số lượng hàng hóa')
plt.title('Biểu đồ số lượng hàng hóa được tạo trong từng ngày')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file JSON
data=pd.read_json("D:/lby/Nam4/HKII/BIGDATA/LeThiBaoYen_Buoi4/hanghoa.json")

# Thêm cột so_luong với giá trị là 1 cho tất cả các sản phẩm
data = data.assign(so_luong=1)

# Tạo groupby theo productlink và tính tổng số lượng sản phẩm
df_grouped = data[data['Active'] == True].groupby(['TenHH'])['Price'].sum().reset_index()



# Vẽ đồ thị cột

fig = px.bar(df_grouped, x='TenHH', y='Price', title='Giá theo sản phẩm',color='TenHH')
fig.update_traces(hovertemplate='Tên trang web: %{x}<br>số lượng: %{y:.0f} sản phẩm')

fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file JSON
data=pd.read_json("D:/lby/Nam4/HKII/BIGDATA/CuoiKy/hanghoa.json")

df_grouped = data.groupby(['TenHH'])['TongTon'].sum().reset_index()

fig = px.bar(df_grouped, x='TenHH', y='TongTon', title='Giá theo sản phẩm',color='TenHH',text=data['TriGiaTon'])
fig.update_traces(hovertemplate='Hàng hóa: %{x}<br>số lượng: %{y:.0f} sản phẩm')

fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Load data from json file
df=pd.read_json("D:/lby/Nam4/HKII/BIGDATA/LeThiBaoYen_Buoi4/hanghoa.json")
df = data.assign(Q=1)

# Group by URL and status, and count the quantity 
df_grouped = df.groupby(['TenHH', 'Active']).sum().reset_index()

# Create bar chart using Plotly
fig = px.bar(df_grouped, x='TenHH', y='Q', color='Active')
fig.show()
```
