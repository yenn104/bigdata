```python
import pandas as pd
import matplotlib.pyplot as plt

# c d liu t file JSON vo DataFrame
df = pd.read_json("D:/lby/Nam4/HKII/BIGDATA/LeThiBaoYen_Buoi4/hanghoa.json")

# Chuyn kiu d liu ca ct CreateDate sang kiu datetime
df['CreatedDate'] = pd.to_datetime(df['CreatedDate'], format='%Y-%m-%d %H:%M:%S.%f')

# Ly s lng hng ha c to trong mi ngy
product_count = df.groupby(df['CreatedDate'].dt.date).size()

# V biu  s lng hng ha c to trong tng ngy
plt.xticks(rotation= 90)

plt.plot(product_count.index, product_count.values)
plt.xlabel('Ngy')
plt.ylabel('S lng hng ha')
plt.title('Biu  s lng hng ha c to trong tng ngy')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# c d liu t file JSON
data=pd.read_json("D:/lby/Nam4/HKII/BIGDATA/LeThiBaoYen_Buoi4/hanghoa.json")

# Thm ct so_luong vi gi tr l 1 cho tt c cc sn phm
data = data.assign(so_luong=1)

# To groupby theo productlink v tnh tng s lng sn phm
df_grouped = data[data['Active'] == True].groupby(['TenHH'])['Price'].sum().reset_index()



# V  th ct

fig = px.bar(df_grouped, x='TenHH', y='Price', title='Gi theo sn phm',color='TenHH')
fig.update_traces(hovertemplate='Tn trang web: %{x}<br>s lng: %{y:.0f} sn phm')

fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# c d liu t file JSON
data=pd.read_json("D:/lby/Nam4/HKII/BIGDATA/CuoiKy/hanghoa.json")

df_grouped = data.groupby(['TenHH'])['TongTon'].sum().reset_index()

fig = px.bar(df_grouped, x='TenHH', y='TongTon', title='Gi theo sn phm',color='TenHH',text=data['TriGiaTon'])
fig.update_traces(hovertemplate='Hng ha: %{x}<br>s lng: %{y:.0f} sn phm')

fig.show()
```
